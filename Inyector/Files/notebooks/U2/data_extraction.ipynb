## **Extraction of data per country - 5 years timeframe**

```python
from pytrends.request import TrendReq

pytrends2 = TrendReq(hl="en-US", tz= '360', timeout=5)

topics_list=['Bitcoin', 'Ether', 'Tether', 'Solana', 'Dogecoin']
pytrends2.build_payload(kw_list = topics_list, timeframe = "today 5-y",  geo = 'MX')
pytrends2 = pytrends2.interest_over_time()

pytrends2 = pytrends2.reset_index()
pytrends2["country"] = 'MX'
new_trends_df2 = pytrends2.melt(
    id_vars = ['date','country'],
    value_vars = topics_list,
    var_name = 'keyword',
    value_name = 'interest'
)
new_trends_df2
```
